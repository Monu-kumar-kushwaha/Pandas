```python
import numpy as np
import pandas as pd
```

```python
--- Merging 2 DataFrames ---
```

```python
employees = pd.DataFrame({
    'employee_id': [1,2,3,4,5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['Hr', 'IT', 'Finance', 'IT', 'Hr']
})

# DataFrame 2 : Salary information
Salaries = pd.DataFrame({
    'employee_id': [1,2,3,6,7],
    'Salary': [60000, 80000, 45000, 67000, 78000],
    'bonus': [1000, 10000, 4500, 5000, 1200]
})
```

```python
employees
```

```python
Salaries
```

```python
pd.merge(employees, Salaries, on = 'employee_id', how = 'inner')
```

```python
pd.merge(employees, Salaries, on = 'employee_id', how = 'outer')
```

```python
pd.merge(employees, Salaries, on = 'employee_id', how = 'left')
```

```python
pd.merge(employees, Salaries, on = 'employee_id', how = 'right')
```

```python
--- Concatination of 2 dataframes ---
```

```python
df1 = pd.DataFrame ({
    'A':['A0', 'A1', 'A2'],
    'B':['B0', 'B1', 'B2'],
    'C':['C0', 'C1', 'C3']
})

df2 = pd.DataFrame ({
    'A':['A3', 'A4', 'A5'],
    'B':['B3', 'B4', 'B5'],
    'C':['C4', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df1,df2], axis = 1)
```

Joining of 2 dataframes

```python
df1 = pd.DataFrame({
    'Name':['Alice', 'Bob', 'Charli']
}, index = [1,2,3])

# second Dataframes
df2 = pd.DataFrame({
    'Scores':[85,90,75]
}, index = [2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
df1.join(df2, how = 'inner')
```

```python
df1.join(df2, how = 'outer')
```

```python
df1.join(df2, how = 'left')
```

```python
df1.join(df2, how = 'right')
```
